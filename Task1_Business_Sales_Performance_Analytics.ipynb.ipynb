```python
import pandas as pd

df = pd.read_csv('dataset/SampleSuperstore.csv')

print(df.head())
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

Remove Duplicate Records

```python
df.drop_duplicates(inplace=True)

print("Dataset Shape:", df.shape)
```

Total Sales

```python
total_sales = df['Sales'].sum()

print("Total Sales =", total_sales)
```

Total Profit

```python
total_profit = df['Profit'].sum()

print("Total Profit =", total_profit)
```

Category-wise Sales

```python
category_sales = df.groupby('Category')['Sales'].sum()

print(category_sales)
```

Chart:

```python
import matplotlib.pyplot as plt

category_sales.plot(kind='bar')
plt.title('Category Wise Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

Region-wise Sales

```python
region_sales = df.groupby('Region')['Sales'].sum()

print(region_sales)
```

Chart:

```python
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Region Wise Sales')
plt.show()
```

Top 10 Products

```python
top_subcategories = df.groupby('Sub-Category')['Sales'].sum()

top_subcategories = top_subcategories.sort_values(ascending=False).head(10)

print(top_subcategories)
```

Chart

```python
import matplotlib.pyplot as plt

top_subcategories.plot(kind='bar')

plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.show()
```

Profit by Category

```python
profit_category = df.groupby('Category')['Profit'].sum()

print(profit_category)

profit_category.plot(kind='bar')
plt.title('Profit by Category')
plt.ylabel('Profit')
plt.show()
```

Sales by Region

```python
region_sales = df.groupby('Region')['Sales'].sum()

print(region_sales)

region_sales.plot(kind='bar')
plt.title('Sales by Region')
plt.ylabel('Sales')
plt.show()
```

Sales by Segment

```python
segment_sales = df.groupby('Segment')['Sales'].sum()

print(segment_sales)
```

Chart
